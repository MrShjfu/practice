# Import Package

```python
from sklearn.cluster import SpectralClustering
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
```

# Load data


```python
digits = load_digits()
image, labels = digits.data, digits.target
```

# Reduce Dimention of image 64 to 2 using PCA 


```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
```

## Result after reduction

```python
train_reduced = pca.fit_transform(image)
plt.grid()
plt.show()
color = ['black','gray','silver','blue','red','brown','teal','olive','yellow','pink']
for i in range(0,9): 
    plt.scatter(train_reduced[labels==i,0], train_reduced[labels ==i,1], s=100, c=color[i], marker='*', label='cluster 1')
```

# Using SpectralClustering to clustering

```python
spectralClustering = SpectralClustering(n_clusters=10, eigen_solver=None, random_state=None, n_init=10, gamma=1.0, affinity='rbf', n_neighbors=10, eigen_tol=0.0, assign_labels='kmeans', degree=3, coef0=1, kernel_params=None, n_jobs=1)
```

```python
result = spectralClustering.fit_predict(train_reduced, y =None)
```

# Visualize

```python
plt.grid()
plt.show()
color = ['black','gray','silver','blue','red','brown','teal','olive','yellow','pink']
for i in range(0,9): 
    plt.scatter(train_reduced[result==i,0], train_reduced[result ==i,1], s=100, c=color[i], marker='*', label='cluster 1')
```
