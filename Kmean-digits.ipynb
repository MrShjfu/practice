# import package

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()
```

## Read data

```python
image, labels = digits.data, digits.target
```

```python
labels
```

## Reduce demonstrate

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
```

```python
train_reduced = pca.fit_transform(image)
plt.grid()
plt.show()
for i in range(0,9): 
    plt.scatter(train_reduced[labels==i,0], train_reduced[labels ==i,1], s=100, c=color[i], marker='*', label='cluster 1')
```

## Using Kmean

```python
km = KMeans(n_clusters=10,init='k-means++',n_init=2,max_iter=200,tol=1e-04,random_state=0)
# max_iter : so lan lap lai thuat toan kmean cho lan chay duy nhat
y_km=km.fit_predict(train_reduced)
```

```python
y_km
```

## Visualize

```python

color = ['black','gray','silver','blue','red','brown','teal','olive','yellow','pink']
plt.grid()
plt.show()
for i in range(0,9): 
    plt.scatter(train_reduced[y_km==i,0], train_reduced[y_km ==i,1], s=100, c=color[i], marker='*', label='cluster 1')
```
